```python
def ldaLearn(X,y):
    # Inputs
    # X - a N x d matrix with each row corresponding to a training example
    # y - a N x 1 column vector indicating the labels for each training example
    #
    # Outputs
    # means - A d x k matrix containing learnt means for each of the k classes
    # covmat - A single d x d learnt covariance matrix 

    # IMPLEMENT THIS METHOD
    Labels = sorted(np.unique(y))
    
    Label_Mean = []
    # calculate the mean for all matrix
    for label in Labels:
        lt = []
        for x, y_iter in zip(X, y):
            if label == y_iter:
                lt.append(x)
        Label_Mean.append(np.mean(lt, axis = 0))

    # Means and Covarience
    means = np.array(Label_Mean).T
    covmat = np.cov(X.T)
    return means, covmat
```

```python
def ldaTest(means,covmat,Xtest,ytest):
    # Inputs
    # means, covmat - parameters of the LDA model
    # Xtest - a N x d matrix with each row corresponding to a test example
    # ytest - a N x 1 column vector indicating the labels for each test example
    # means - A d x k matrix containing learnt means for each of the k classes
    # covmat - A single d x d learnt covariance matrix 
    # Outputs
    # acc - A scalar accuracy value
    # ypred - N x 1 column vector indicating the predicted labels
    # https://web.stanford.edu/class/stats202/content/lec9.pdf
    
    
    y_estimate = []
    label_list = np.array([1,2,3,4,5])
    
    for x in Xtest:
        x = x.reshape(1, 2)
        label_prob = [0] * len(label_list)
        for index,item in  enumerate(label_list):
            u = means[:,index].reshape (1,2)
            mu = u
            sigma = covmat
            Invsigma = inv(sigma)
            power_term =  -1 / 2 * np.matmul((x-u), np.matmul(Invsigma, (x-u).T))
            exp_power_term  = math.exp(power_term)
            label_prob[index] =  ( 1/ ((2*math.pi)**(x.shape[1]/2))* (np.linalg.det(sigma)**(1/2)))*exp_power_term
            #print (label_prob[index])
        y_estimate.append(label_list[label_prob.index(max(label_prob))])
    
    # Scalar Accuracy
    acc_count = 0
    for y_est,actual_y in zip(y_estimate, ytest):
        if y_est == actual_y:
            acc_count += 1

    acc = acc_count / len(y_estimate)
    y_estimate = np.array(y_estimate)
    return acc, y_estimate
```

```python
def qdaTest(means,covmats,Xtest,ytest):
    # Inputs
    # means, covmats - parameters of the QDA model
    # Xtest - a N x d matrix with each row corresponding to a test example
    # ytest - a N x 1 column vector indicating the labels for each test example
    # Outputs
    # acc - A scalar accuracy value
    # ypred - N x 1 column vector indicating the predicted labels
    # https://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.det.html
    # IMPLEMENT THIS METHOD

    y_estimate = []
    label_list = np.array([1,2,3,4,5])

    for x in Xtest:
        x = x.reshape(1, 2)
        label_prob = [0] * len(label_list)
        for index,item in  enumerate(label_list):
            u = means[:,index].reshape (1,2)
            sigma = covmats[index]
            Invsigma = inv(sigma)
            power_term =  -1 / 2 * np.matmul((x-u), np.matmul(Invsigma, (x-u).T))
            exp_power_term  = math.exp(power_term)
            label_prob[index] =  ( 1/ ((2*math.pi)**(x.shape[1]/2))* (np.linalg.det(sigma)**(1/2)))*exp_power_term
            #print (label_prob[index])
        y_estimate.append(label_list[label_prob.index(max(label_prob))])

    #Scalar Accuracy
    count = 0
    for y_est, y_tst in zip(y_estimate, ytest):
        if y_est == y_tst:
            count += 1

    acc = count / len(y_estimate)
    y_esti = np.array(y_estimate)
    return acc, y_esti
```

```python
def learnOLERegression(X,y):
    # Inputs:                                                                   
    # X = N x d                                                                 
    # y = N x 1                                                                 
    # Output:                                                                   
    # w = d x 1                                                                 

    # IMPLEMENT THIS METHOD  
    # w = (xTx)^1 *Xty

    term1 = np.dot(np.transpose(X),X)
    term2 = np.dot(np.transpose(X),y)
    iterm1 = inv(term1)
    w = np.dot(iterm1,term2)
    return w
```

```python
def learnRidgeRegression(X,y,lambd):
    # Inputs:                                                                   
    # X = N x d                                                                 
    # y = N x 1 
    # lambd = ridge parameter (scalar)
    # Output:                                                                   
    # w = d x 1                                                                 
    # MAP = inv (lambda * ID + XT X) Xt y

    # IMPLEMENT THIS METHOD 
    term1b = np.dot(np.transpose(X),X)
    term1a = lambd*np.eye(term1b.shape[0])
    term2 = np.dot(np.transpose(X),y)
    w = np.dot(inv(term1b+term1a), term2)
    return w
```

```python
def testOLERegression(w,Xtest,ytest):
    # Inputs:
    # w = d x 1
    # Xtest = N x d
    # ytest = X x 1
    # Output:
    # mse
  
    # IMPLEMENT THIS METHOD
    y_est = np.dot(Xtest,w)
    error = y_est - ytest
    #https://mathinsight.org/dot_product_matrix_notation
    mse = np.dot(np.transpose(error),error)/(error.shape[0])
    return mse
```

```python
def regressionObjVal(w, X, y, lambd):
    # compute squared error (scalar) and gradient of squared error with respect
    # to w (vector) for the given data X and y and the regularization parameter
    # lambda                                                                  
    
    # IMPLEMENT THIS METHOD  
    
    y_estimate = np.dot(X,np.transpose(w))
    temp  = np.zeros(y.shape[0])
    error_list = list()
    i = 0;
    for y_est,y_out in zip(y_estimate,y):
        temp[i] = y_est - y_out;
        i = i+1
    
    error = np.dot(np.transpose(temp),temp)/temp.shape[0]
    gradient = np.dot(np.transpose(X), temp)/X.shape[0] + (lambd/X.shape[0]) * w
    return error, gradient
```

```python
def mapNonLinear(x,p):
    # Inputs:                                                                   
    # x - a single column vector (N x 1)                                        
    # p - integer (>= 0)                                                        
    # Outputs:                                                                  
    # Xd - (N x (d+1))                                                         
    # 1 x x2 x^p
    # IMPLEMENT THIS METHOD
    N  = x.shape[0]
    d  = p + 1
    Xd = np.zeros((N,d))
    # Iterate per colum and take power of x  with the index. yielding 1,x,x^2,x^3... x^p
    for column in range(p+1):
        Xd[:,column] = pow(x,column)
        
    return Xd
```
